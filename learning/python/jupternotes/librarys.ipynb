```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = [10,20,30,40,50]

# print("Dataset:")
# print(data.head())

mean_value = np.mean(data)
print('Mean:', mean_value)

median_value = np.median(data)
print('Median:', median_value)


# mode_value = mode_result.mode[1]
# print('Mode:', mode_value)

std_dev = np.std(data)
print('Standard Deviation:', std_dev)


variance = np.var(data)
print('Variance:', variance)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


x = [10, 1, 2, 3, 4]
y = [10, 1, 4, 9, 16]

df = pd.DataFrame({'x': x, 'y': y})


plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['y'], linestyle='-', marker='o', color='b', label='y = x^2')
plt.title('Line Plot: y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# sns.barplot

```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['x'], color='purple', alpha=0.7, bins=10) 
plt.title("Histogram: Distribution of x values")
plt.xlabel('x')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['y'])
plt.title('Box Plot: y = x^2')
plt.ylabel('y')
plt.show()
```

```python
categories = ['A' , 'B', 'C', 'D', 'E']
values = [1,2,3,2,1]

plt.figure(figsize=(8,40))
plt.pie(values, labels = categories, autopct = '%1.1f%%', colors = ['#ff0000' , '#0000ff' , '#3cb371' , '#ee82ee' , '#ffa500'])
plt.title('Pie Chart: Distribution of Categories')
plt.show()

```
